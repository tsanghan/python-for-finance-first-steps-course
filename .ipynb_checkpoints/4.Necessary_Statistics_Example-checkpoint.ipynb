## Example-1

Let's extract a stock price from `yahoo finance`, calculate stock return, visualize it and decide which distribution it fits.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn')
```

```python
stock = 'MMM'
start = datetime.datetime(2000,1,1)
end = datetime.datetime(2020,1,1)
MMM = yf.download(stock,start=start,end = end, interval='1d')
```

```python
plt.plot(MMM.Open)
plt.title('Open Price of 3M')
plt.xlabel('Date')
plt.ylabel('$')
plt.show()
```

```python
MMM.describe()
```

```python
plt.hist(MMM.Close)
plt.show()
```

```python
plt.boxplot(MMM.Close)
plt.show()
```

```python
MMM_return=MMM.Close.pct_change().dropna()
```

```python
plt.plot(MMM_return)
plt.title('Return of 3M')
plt.xlabel('Date')
plt.ylabel('$')
plt.show()
```

```python
MMM_return.hist()
plt.show()
```

```python
from scipy import stats
```

```python
k2, p = stats.normaltest(MMM_return)
print("The Normal test statistic is {} and p-value is {}".format(k2,p))
```

This test accepts the null hypothesis stating that the sample does not come from a normal distribution.

```python
MMM_return_sorted=MMM_return.sort_values()
```

```python
import scipy.stats as stats
import pylab
stats.probplot(MMM_return_sorted,dist="norm",plot=pylab)
pylab.show()
```

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')
data = np.array(MMM_return).reshape(-1, 1)
pt.fit(data)
transformed_data = pt.transform(data)
```

```python
transformed_k2, transformed_p = stats.normaltest(transformed_data)
print("The Normal test statistic is {} and p-value is {}".format(transformed_k2,transformed_p))
```

## Example-2

What do you think about the relationship between S&P-500 index return and GDP growth?


```python

import datetime
import yfinance as yf

stock = '^GSPC'
start = datetime.datetime(1990,1,1)
end = datetime.datetime(2020,1,1)
SP = yf.download(stock,start=start,end = end, interval='1d')
```

```python
SP.Close.plot()
plt.xlabel('Date')
plt.ylabel('$')
plt.show()
```

```python
SP['year']=pd.DatetimeIndex(SP.index).year
```

```python
SP
```

```python
SP_annual=SP.Close.resample('Y').mean()
```

```python
SP_annual_return=SP_annual.pct_change().dropna()
```

```python
plt.plot(SP_annual_return)
plt.xlabel('Date')
plt.ylabel('$')
plt.title('S&P-500 Annual Return')
plt.show()
```
