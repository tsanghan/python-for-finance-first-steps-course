If we change the code in our cell or run it multiple times, we may have changed the variables or output in an undesired way. The same thing can also happen if we run cells in the wrong order. By convention `np` is used to represent numpy. Now, we are ready to dig into numpy. As we mentioned above, NumPy is a basic package for storing and modifying mathematical data in Python. NumPy does this with a new data structure called `array`.

One of the main advantages of the array is the multidimensionality property and simply by typing `np.array()`, we can create a list.

```python
multi = np.array([[0, 1, 2, 3],[4, 5, 6, 7]])
multi
```

It is also possible to make some changes in the array. Let's assign different the elements in an array starting from the second index:

```python
array = np.array([0, 1, 2, 3])
array[2:] = [5,4]
print(array)
```

As we mentioned, there are many operations we can do with array. `append` is just one of them. It basically add new elements to the array.

```python
np.append(array, (7,8,9))
```

Finally, I would like to talk about basic mathematical operations that we can do in numpy. These operations include summation, multiplication, division, element-wise operations, and aggregator function. Let's first start with summation, multiplication, and division.

```python
array1 = np.array([0, 1, 2, 3])
array2 = np.array([2, 4, 6, 8])
print (array1 + array2)
```

```python
array1 = np.array([[0, 1, 2, 3],[4, 5, 6, 7]])

print(array2 = array1*2)

print(array2)
```

```python
print(array1/array2)
```

Thus, the application of basic mathematical operations are easy-to-apply. Now, it is time to  do some element-wise operations. Among others, these operations include taking squared and square root. To do that we will use `.square()`and `.sqrt()` functions.

```python
x = np.array(np.arange(5))
np.square(x)
```

```python
np.sqrt(x)
```

As a final step, aggregator functions in numpy is introduced. Finding minimum, maximum, mean values, and standard deviation are some examples of aggregator functions.

```python
age=np.array([18, 25, 22, 34, 26, 27, 19, 28, 30, 24])
```

```python
print(np.max(age))
print(np.min(age))
print(np.mean(age))
print(np.std(age))
```

So, it turns out finding summary statistics about a data is quite easy in numpy. For those who wants to learn more can apply to this [website](https://numpy.org/doc/stable/).

### 2.4.3. Histogram

The histogram is a great tool to show the possible values of a variable and how frequent these values are. In addition to that, histogram gives us an initial idea about the distribution of the data has. Knowing the distribution of a data may be quite important to know before running a model because some models assume certain type of distribution.

The histogram begins by dividing the values into bins. These bins are equal-sized ranges with an upper and lower limit. The histogram then counts how many values are in each bin and draws a bar corresponding to this number for each bin. This gives us an idea of the frequency of the range of its variable. So, histograms are great for visualizing the distribution of a variable.

Here, we draw the histogram of `total_profit` and `total_units` variables.

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(df.total_profit)
plt.title('Total Profit')
plt.subplot(1, 2, 2)
plt.hist(df.moisturizer)
plt.title('Moisturizer')

plt.show()
```

The first one occurs when moisturizer takes on values around 1200 and the other happens when it takes on values around 1800.

### 2.4.4 Box Plot

Another useful visualization is the boxplot. We can create a boxplot with the `.boxplot()` function. Boxplot is a relatively simple visualization. Let's create one and then examine its properties. 

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(df.facewash)
plt.ylabel('Face Wash')

plt.subplot(1, 2, 2)
plt.boxplot(df.bathingsoap)
plt.ylabel('Bathing Soap')

plt.tight_layout()
plt.show()
```

The boxplot consists of a box containing what is called `quartile` (interquartile range or IQR for short). First of all, let's explain what the quartile means. We can divide a variable into four quartiles based on the values it contains:

* From the lowest value to 25%,
* From 25% to the median,
* From the median to 75% and
* From 75% to the maximum value.

These groups are called quartiles and the values that the two in the middle quartiles cover  are called `interquartile range`. Thus, the box we see in the boxplot contains all the data from 25% to 75% and shows the `median` with a red line.

There are vertical lines called whiskers above and below the box (for this reason, the boxplot is sometimes called "box and whisker diagrams"). The whiskers extends up and down 1.5x from the top and bottom of the IQR box. If all values are in this range, the whiskers goes up to the maximum and minimum values of the variable. Any value beyond 1.5 times the IQR represents as a tick beyond the end of the whiskers. For example, in the line above on the left hand side, there are some values above the whiskers.

Boxplot is not useful for learning about the distribution of a variable. However, it is very useful for eyeballing the range of values and for detecting possible `outliers`. Data points marked with a dot or circle on a box diagram are generally considered to be outliers. Therefore, box diagrams are widely used when working on new data. For more information about Boxplot, see the `.boxplot()` [documentation](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.boxplot).
