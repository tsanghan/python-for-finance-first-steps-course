## Example-1

As an example, let us calculate the length of the hypotenuse of a right triangle, whose perpendicular edges are given as parameters, by writing a function where we name it `hip`:

```python
def hip(edge_1, edge_2):
    hypotenuse = ((edge_1**2) + (edge_2**2))**0.5
    return hypotenuse
```

```python
print(hip(3, 4))
print(hip(6, 8))
print(hip(5, 12))
```

As another example, let us write a function that lists the square of counting numbers less than a number entered:

```python
def little_square_to_list(x):
    little_list=[]
    
    for y in range(1, x):
        little_list.append(y**2)
    
    return little_list

print(little_square_to_list(15))
```

Define a function which can generate and print a list where the values are square of numbers between 1 and 20 (both included).

```python
def printList():
    li=list()
    for i in range(1,21):
        li.append(i**2)
    print(li)
```

```python
printList()
```

## Example-2

Let's find the prime numbers between 1 and 100.

```python
sum_of_primes=0

for x in range(1,101):
    count=0
    
    if x<=2:
        sum_of_primes+=x
        
    elif x%2!=0:
        for y in range(2,x//2):
            if (x%y)==0:
                count+=1
                
        if count==0:
            sum_of_primes+=x
            
print(sum_of_primes)
```

## Example-3

What do you think about the relationship between S&P-500 index return and GDP growth?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import datetime
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn')
```

```python
stock = '^GSPC'
start = datetime.datetime(1990,1,1)
end = datetime.datetime(2020,1,1)
SP = yf.download(stock,start=start,end = end, interval='1d')
```

```python
SP.Close.plot()
plt.xlabel('Date')
plt.ylabel('$')
plt.show()
```

```python
SP['year']=pd.DatetimeIndex(SP.index).year
```

```python
SP
```

```python
SP_annual=SP.Close.resample('Y').mean()
```

```python
SP_annual_return=SP_annual.pct_change().dropna()
```

```python
plt.plot(SP_annual_return)
plt.xlabel('Date')
plt.ylabel('$')
plt.title('S&P-500 Annual Return')
plt.show()
```
